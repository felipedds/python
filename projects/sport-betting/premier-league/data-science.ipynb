```python
import numpy as np
import pandas as pd

pd.set_option('display.min_rows', 200)
pd.set_option("display.max_rows", 600)
pd.set_option("display.max_columns", 100)

df = pd.read_csv("https://www.football-data.co.uk/mmz4281/2223/E0.csv")

df.rename(columns={"Date":"date", "Time":"time", "FTHG":"full_time_home_goals", "FTAG":"full_time_away_goals", "FTR":"full_time_result", 
                   "HomeTeam":"home_team", "AwayTeam":"away_team", "HTHG":"half_time_home_team_goals", 
                   "HTAG":"half_time_away_team_goals", "HTR":"half_time_result", "HS":"home_team_shots", "AS":"away_team_shots", 
                   "HST":"home_team_shots_on_target", "AST":"away_team_shots_on_target"}, inplace=True)
df = df.astype({"date":"datetime64"})
df = df.convert_dtypes(infer_objects=True)
#df = df[["date", "home_team", "away_team", "full_time_home_goals", "full_time_away_goals", "full_time_result", "B365H", "B365D", "B365A", "PSH", "PSD", "PSA"]]
df.tail()
```

<h5>Mean odds home/draw/away per Team - BET365</h5>

```python
odds_home_team_mean = df.groupby(df["home_team"])["B365H"].mean()
odds_draw_team_mean = df.groupby(df["home_team"])["B365D"].mean()
odds_away_team_mean = df.groupby(df["away_team"])["B365A"].mean()

odds_mean = pd.concat([odds_home_team_mean, odds_draw_team_mean, odds_away_team_mean], axis=1)
odds_mean
```

<h5>Mean goals home/away per Team</h5>

```python
goals_home_team_mean = df.groupby(df["home_team"])["full_time_home_goals"].mean()
goals_away_team_mean = df.groupby(df["away_team"])["full_time_away_goals"].mean()
goals_team_mean = (goals_home_team_mean + goals_away_team_mean) / 2

goals_mean = pd.concat([goals_home_team_mean, goals_away_team_mean, goals_team_mean], axis=1, keys = ['goals_home_team_mean', 'goals_away_team_mean', 'goals_team_mean']).round(3)
goals_mean
```

```python
def season_results(team):
    # Filter
    home_team = df.loc[df["home_team"] == team]
    away_team = df.loc[df["away_team"] == team]

    # Goals scored sum
    home_team_goals_scored_sum = home_team["full_time_home_goals"].sum()
    away_team_goals_scored_sum = away_team["full_time_away_goals"].sum()
    goals_scored = home_team_goals_scored_sum + away_team_goals_scored_sum

    # Goals conceded sum
    home_team_goals_conceded_sum = home_team["full_time_away_goals"].sum()
    away_team_goals_conceded_sum = away_team["full_time_home_goals"].sum()
    goals_conceded = home_team_goals_conceded_sum + away_team_goals_conceded_sum

    # Goals scored mean
    home_team_goals_scored_mean = home_team["full_time_home_goals"].mean()
    away_team_goals_scored_mean = away_team["full_time_away_goals"].mean()

    # Goals conceded mean
    home_team_goals_conceded_mean = home_team["full_time_away_goals"].mean()
    away_team_goals_conceded_mean = away_team["full_time_home_goals"].mean()

    print(f"The {team} have scored a total of {goals_scored} goals in this season.")
    print(f"The {team} have conceded score a total of {goals_conceded} goals in this season.")
    print(f"The {team} have scored a total of {home_team_goals_scored_sum} goals in this home.")
    print(f"The {team} have scored a total of {away_team_goals_scored_sum} away goals.")
    print(f"The {team} have conceded score a total of {home_team_goals_conceded_sum} goals in this home.")
    print(f"The {team} have conceded score a total of {away_team_goals_conceded_sum} away goals.")
    print(f"The {team} have a mean of score {home_team_goals_scored_mean.round(2)} goals in this home.")
    print(f"The {team} have a mean of score {away_team_goals_scored_mean.round(2)} away goals.")
    print(f"The {team} have a mean of conceded score of {home_team_goals_conceded_mean.round(2)} goals in this home.")
    print(f"The {team} have a mean of conceded score of {away_team_goals_conceded_mean.round(2)} away goals.")

team = "Aston Villa"
season_results(team)
```
