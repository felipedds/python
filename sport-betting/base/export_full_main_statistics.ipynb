```python
# Points done at home
def points_home(result):
    if result == "H":
        points = 3
    elif result == "D":
        points = 1
    else:
        points = 0
    return points

# Points done away
def points_away(result):
    if result == "A":
        points = 3
    elif result == "D":
        points = 1
    else:
        points = 0
    return points

df["points_home"] = df.apply(lambda row: points_home(row["result_ft"]), axis=1)
df["points_away"] = df.apply(lambda row: points_away(row["result_ft"]), axis=1)
```

```python
# Informations about Goals
df["total_goals_ht"] = df.apply(lambda row: (row["home_goals_ht"] + row["away_goals_ht"]), axis=1)
df["total_goals_ft"] = df.apply(lambda row: (row["home_goals_ft"] + row["away_goals_ft"]), axis=1)
df["over05_ht"] = df.apply(lambda row: 1 if (row["total_goals_ht"] > 0 ) else 0, axis=1)
df["over05_ft"] = df.apply(lambda row: 1 if (row["total_goals_ft"] > 0 ) else 0, axis=1)
df["over15_ft"] = df.apply(lambda row: 1 if (row["total_goals_ft"] > 1 ) else 0, axis=1)
df["over25_ft"] = df.apply(lambda row: 1 if (row["total_goals_ft"] > 2 ) else 0, axis=1)
df["both_teams_score"] = df.apply(lambda row: 1 if (row["home_goals_ft"] > 0 and row["away_goals_ft"] > 0) else 0, axis=1)

# Percentages
df["percentage_over05_ht_home"] = (df.groupby("home")["over05_ht"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)
df["percentage_over05_ht_away"] = (df.groupby("away")["over05_ht"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)

df["percentage_over05_ft_away"] = (df.groupby("home")["over05_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)
df["percentage_over05_ft_away"] = (df.groupby("away")["over05_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)

df["percentage_over15_ft_home"] = (df.groupby("home")["over15_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)
df["percentage_over15_ft_away"] = (df.groupby("away")["over15_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)

df["percentage_over25_ft_home"] = (df.groupby("home")["over25_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)
df["percentage_over25_ft_away"] = (df.groupby("away")["over25_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)

df["percentage_both_teams_score_home"] = (df.groupby("home")["both_teams_score"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)
df["percentage_both_teams_score_away"] = (df.groupby("away")["both_teams_score"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True) * 100).round(2)

# Statistics of Points (Mean, Standard deviation and Coefficient of variation)
df["mean_points_home"] = (df.groupby("home")["points_home"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True)).round(2)
df["mean_points_away"] = (df.groupby("away")["points_away"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True)).round(2)

df["standard_deviation_points_home"] = (df.groupby("home")["points_home"].rolling(window=5, min_periods=1).std().reset_index(0, drop=True)).round(2)
df["standard_deviation_points_away"] = (df.groupby("away")["points_away"].rolling(window=5, min_periods=1).std().reset_index(0, drop=True)).round(2)

df["coefficient_variation_points_home"] = (df["standard_deviation_points_home"] / df["mean_points_home"]).round(2)
df["coefficient_variation_points_away"] = (df["standard_deviation_points_away"] / df["mean_points_away"]).round(2)

# Statistics of Goals (Mean, Standard deviation and Coefficient of variation)
df["mean_goals_scored_home"] = (df.groupby("home")["home_goals_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True)).round(2)
df["mean_goals_conceded_home"] = (df.groupby("home")["away_goals_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True)).round(2)
df["mean_goals_scored_away"] = (df.groupby("away")["away_goals_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True)).round(2)
df["mean_goals_conceded_away"] = (df.groupby("away")["home_goals_ft"].rolling(window=5, min_periods=1).mean().reset_index(0, drop=True)).round(2)

df["standard_deviation_goals_scored_home"] = (df.groupby("home")["home_goals_ft"].rolling(window=5, min_periods=1).std().reset_index(0, drop=True)).round(2)
df["standard_deviation_goals_conceded_home"] = (df.groupby("home")["away_goals_ft"].rolling(window=5, min_periods=1).std().reset_index(0, drop=True)).round(2)
df["standard_deviation_goals_scored_away"] = (df.groupby("away")["away_goals_ft"].rolling(window=5, min_periods=1).std().reset_index(0, drop=True)).round(2)
df["standard_deviation_goals_conceded_away"] = (df.groupby("away")["home_goals_ft"].rolling(window=5, min_periods=1).std().reset_index(0, drop=True)).round(2)

df["coefficient_variation_goals_scored_home"] = (df["standard_deviation_goals_scored_home"] / df["mean_goals_scored_home"]).round(2).fillna(0)
df["coefficient_variation_goals_conceded_home"] = (df["standard_deviation_goals_conceded_home"] / df["mean_goals_conceded_home"]).round(2).fillna(0)
df["coefficient_variation_goals_scored_away"] = (df["standard_deviation_goals_scored_away"] / df["mean_goals_scored_away"]).round(2).fillna(0)
df["coefficient_variation_goals_conceded_away"] = (df["standard_deviation_goals_conceded_away"] / df["mean_goals_conceded_away"]).round(2).fillna(0)

# Probability (Odd's Normalization)
df["probability_home"] = df.apply(lambda row: (1 / row["home_odds"]), axis=1).round(2)
df["probability_draw"] = df.apply(lambda row: (1 / row["draw_odds"]), axis=1).round(2)
df["probability_away"] = df.apply(lambda row: (1 / row["away_odds"]), axis=1).round(2)

df.dropna(inplace=True)
df.sort_values("date")
df.reset_index(inplace=True, drop=True)
df.to_csv("full_main_statistics.csv")
df.tail()
```
