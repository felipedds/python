<h5>Import dataset</h5>
<h6>Load the dataset from a file .csv<br>
Font: https://www.footballwebpages.co.uk/premier-league/league-table<h6>

```python
import pandas as pd

df = pd.read_csv("/home/felipe/Desktop/python/projects/sport-betting/premier-league/base/england-premier-league-table-2022-2023.csv")
df.drop(columns={"Unnamed: 0"}, inplace=True)
df.rename(columns={"Unnamed: 1":"team", "P":"played_home", "W":"won_home", "D":"draw_home", "L":"loss_home", 
                   "P.1":"played_away", "W.1":"won_away", "D.1":"draw_away", "L.1":"loss_away",
                   "P.2":"played_total", "W.2":"won_total", "D.2":"draw_total", "L.2":"loss_total", 
                   "F":"goals_for", "A":"goals_against", "+/-":"goal_difference", "Pts":"points"}, inplace=True)
df = df.sort_values(by="points", ascending=False)
df.tail()
```

<h4>Odds to Probability / Probability to Odds</h4>
<h6>Converting Odds to Probability and Probability to Odds</h6>

```python
odd = 1.88
odd_to_probability = 100 / odd
print(f"The probability is: {odd_to_probability}")

# odd to probability
# 0.05 = 0.05 × 100% = 5%
# 0.3 = 0.3 × 100% = 30%
# 0.35 = 0.35 × 100% = 35%
# 3.5 = 3.5 × 100% = 350%

probability = odd_to_probability / 100  # 53.19% 
probability_to_odd = (1 / probability)
print(f"The odd is: {probability_to_odd}")
```

<h5>Feature Engeneering<h5>
<h6>Invent new features<h6>

```python
# Feature Engeneering
df["winning_percentage"] = (df["won_total"] / df["played_total"]).round(decimals=3) * 100
df["draw_percentage"] = (df["draw_total"] / df["played_total"]).round(decimals=3) * 100
df["loss_percentage"] = (df["loss_total"] / df["played_total"]).round(decimals=3) * 100
df.tail()
```
